```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df=pd.read_csv(r'C:\Users\sangeeth kumar\PycharmProjects\python_class2\Sangeeth_Sessions\cereal.csv')
```

```python
y=df['rating']
```

```python
df.dtypes
```

```python
cat=df.select_dtypes(include="object")
num=df.select_dtypes(include="number")
```

```python
num.head()
```

```python
cat=pd.get_dummies(cat,dtype='int')
print(cat.head(20))
x=pd.concat([cat,num],axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
print(x_train.shape, x_test.shape,y_train.shape,y_test.shape)
```

```python
model=LinearRegression(normalize=True).fit(x_train,y_train)
```

```python
train_preds=model.predict(x_train)
```

```python
print("Train Mean Squared Error:",mean_squared_error(y_train,train_preds))
print("Train R2 value:",r2_score(y_train,train_preds))
```

```python
predictions=model.predict(x_test)
print(predictions)
```

```python
print("Mean Squared Error:",mean_squared_error(y_test,predictions))
```

```python
print("R2 value:",r2_score(y_test,predictions))
```

```python
r_model=Ridge(alpha=0.01).fit(x_train,y_train)
ridge_predictions=r_model.predict(x_test)
```
